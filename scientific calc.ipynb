```python
import ipywidgets as w
import math

# Input and Output Boxes
h = w.Text()
output = w.FloatText()

# Button creation with colors and styles
# First row
buttonRad = w.Button(description="Rad", style={'button_color': 'lightblue'})
buttonDeg = w.Button(description="Deg", style={'button_color': 'lightblue'})
buttonFact = w.Button(description="x!", style={'button_color': 'lightblue'})
buttonopen = w.Button(description="(",style={'button_color': 'lightgray'})
buttonclose = w.Button(description=")",style={'button_color': 'lightgray'})
buttonmod = w.Button(description="%",style={'button_color': 'lightgray'})
buttonAC = w.Button(description="AC",style={'button_color': 'red'})

# Second row of Boxes
buttonInv = w.Button(description="Inv", style={'button_color': 'lightblue'})
buttonSin = w.Button(description="Sin",  style={'button_color': 'lightblue'})
buttonLn = w.Button(description="Ln", style={'button_color': 'lightblue'})
button7 = w.Button(description="7", style={'button_color': 'lightgray'})
button9 = w.Button(description="9", style={'button_color': 'lightgray'})
button8 = w.Button(description="8", style={'button_color': 'lightgray'})
buttondiv = w.Button(description="/", style={'button_color': 'orange'})

# Third row of Boxes
buttonPie = w.Button(description="π",  style={'button_color': 'lightblue'})
buttonCos = w.Button(description="Cos",  style={'button_color': 'lightblue'})
buttonLog = w.Button(description="Log", style={'button_color': 'lightblue'})
button4 = w.Button(description="4", style={'button_color': 'lightgray'})
button5 = w.Button(description="5", style={'button_color': 'lightgray'})
button6 = w.Button(description="6", style={'button_color': 'lightgray'})
buttonmul = w.Button(description="x", style={'button_color': 'orange'})

# Fourth row of Boxes
buttone = w.Button(description="e",  style={'button_color': 'lightblue'})
buttonTan = w.Button(description="Tan", style={'button_color': 'lightblue'})
buttonroot = w.Button(description="√", style={'button_color': 'lightblue'})
button1 = w.Button(description="1", style={'button_color': 'lightgray'})
button2 = w.Button(description="2", style={'button_color': 'lightgray'})
button3 = w.Button(description="3", style={'button_color': 'lightgray'})
buttonminus = w.Button(description="-", style={'button_color': 'orange'})

# Fifth row of Boxes
buttonAns = w.Button(description="Ans",  style={'button_color': 'lightblue'})
button10power = w.Button(description="10^x", style={'button_color': 'lightblue'})
buttonpower = w.Button(description="x^y", style={'button_color': 'lightblue'})
button0 = w.Button(description="0", style={'button_color': 'lightgray'})
buttondot = w.Button(description=".", style={'button_color': 'lightgray'})
buttonequal = w.Button(description="=", style={'button_color': 'red'})
buttonplus = w.Button(description="+", style={'button_color': 'orange'})

# Backspace Button
buttonBS = w.Button(description="BackSpace", style={'button_color': 'red'})

# Layout for the buttons
first_box = w.HBox([buttonRad, buttonDeg, buttonFact, buttonopen, buttonclose, buttonmod, buttonAC])
second_box = w.HBox([buttonInv, buttonSin, buttonLn, button7, button8, button9, buttondiv])
third_box = w.HBox([buttonPie, buttonCos, buttonLog, button4, button5, button6, buttonmul])
fourth_box = w.HBox([buttone, buttonTan, buttonroot, button1, button2, button3, buttonminus])
fifth_box = w.HBox([buttonAns, button10power, buttonpower, button0, buttondot, buttonequal, buttonplus])
out_box = w.HBox([output, buttonBS])

# Assembling the layout
box = w.VBox([h, first_box, second_box, third_box, fourth_box, fifth_box, out_box])
tab = w.Tab([box])
tab.set_title(0, "Calculator")
display(tab)

# Functions of 1st row Buttons
def fRad(self):
    h.value = h.value + "radians("

def fDeg(self):
    h.value = h.value + "degrees("

def fFact(self):
    h.value = h.value + "factorial("

def fopen(self):
    h.value = h.value + "("

def fclose(self):
    h.value = h.value + ")"

def fmod(self):
    h.value = h.value + "%"

def fAC(self):
    h.value = ""
    output.value = ""

# Functions of 2nd row buttons
def fInv(self):
    h.value = h.value + "1/("

def fSin(self):
    h.value = h.value + "math.sin("  # Update to use math.sin

def ln(self):
    h.value = h.value + "math.log("  # Update to use math.log

def f7(self):
    h.value = h.value + "7"

def f8(self):
    h.value = h.value + "8"

def f9(self):
    h.value = h.value + "9"

def fdiv(self):
    h.value = h.value + "/"

# Functions of 3rd row buttons
def fpie(self):
    h.value = h.value + "(22/7)"

def fCos(self):
    h.value = h.value + "math.cos("  # Update to use math.cos

def f4(self):
    h.value = h.value + "4"

def f5(self):
    h.value = h.value + "5"

def f6(self):
    h.value = h.value + "6"

def fmul(self):
    h.value = h.value + "*"

# Functions of 4th row buttons
def fe(self):
    h.value = h.value + ("2.71828**")

def fTan(self):
    h.value = h.value + "math.tan("  # Update to use math.tan

def square_root(self):
    h.value = h.value + "math.sqrt("  # Update to use math.sqrt

def f1(self):
    h.value = h.value + "1"

def f2(self):
    h.value = h.value + "2"

def f3(self):
    h.value = h.value + "3"

def fminus(self):
    h.value = h.value + "-"

# Functions of 5th row buttons
def f10power(self):
    h.value = h.value + "(10**"

def fpower(self):
    h.value = h.value + "**"

def f0(self):
    h.value = h.value + "0"

def fdot(self):
    h.value = h.value + "."

def fequal(self):
    s = eval(h.value)  # Use eval to calculate the expression
    output.value = s

def fplus(self):
    h.value = h.value + "+"

def fbackspace(self):
    h.value = h.value[0:-1]

# Row 1 function calling by onclick
buttonRad.on_click(fRad)
buttonDeg.on_click(fDeg)
buttonFact.on_click(fFact)
buttonopen.on_click(fopen)
buttonclose.on_click(fclose)
buttonmod.on_click(fmod)
buttonAC.on_click(fAC)

# Row 2 function calling by onclick
buttonInv.on_click(fInv)
buttonSin.on_click(fSin)
buttonLn.on_click(ln)
button7.on_click(f7)
button8.on_click(f8)
button9.on_click(f9)
buttondiv.on_click(fdiv)

# Row 3 function calling by onclick
buttonPie.on_click(fpie)
buttonCos.on_click(fCos)
buttonLog.on_click(ln)
button4.on_click(f4)
button5.on_click(f5)
button6.on_click(f6)
buttonmul.on_click(fmul)

# Row 4 function calling by onclick
buttone.on_click(fe)
buttonTan.on_click(fTan)
buttonroot.on_click(square_root)
button1.on_click(f1)
button2.on_click(f2)
button3.on_click(f3)
buttonminus.on_click(fminus)

# Row 5 function calling by onclick
buttonAns.on_click(fequal)
button10power.on_click(f10power)
buttonpower.on_click(fpower)
button0.on_click(f0)
buttondot.on_click(fdot)
buttonequal.on_click(fequal)
buttonplus.on_click(fplus)
buttonBS.on_click(fbackspace)
```
